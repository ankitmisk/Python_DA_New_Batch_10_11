```python
import random
rn = random.randint(1,100)
even_10 = [rn for i in range(10) if rn%2==0 ]
print(even_10)
```

```python
# random.randint()
random.seed(2345)
result = []

for i in range(234):
    rn = random.randint(1,100)
    if (rn not in result) and rn%2==0:
        result.append(rn)
    
    if len(result) == 10:
        break
        
print(result)
```
