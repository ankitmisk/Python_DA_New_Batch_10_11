```python
import pymysql as py
def make_connection():
    try:
        conn = py.connect(user = 'root',host='localhost',password='1401',autocommit=True)
        cur = conn.cursor()
        
        return cur
    except:
        return False
```

```python
make_connection()
```

```python
cur = make_connection()
if not cur:
    print('Connection Failed!!')
else:
    print('Connection established!!')
```

```python
import pandas as pd
def execute_query(query):
    try:
        cur.execute(query)

        result = cur.fetchall()

        if len(result)>0:
            all_columns = []
            
            for i in cur.description:
                all_columns.append(i[0])
                
            table = pd.DataFrame(result,columns=all_columns)
            
            return table
            
        else:
            return 'Query executed successfully!!'
        
    except Exception as e:
        print('Error❌',e)
```

```python
execute_query('select * from world.country;')
```

```python
import random
import time
full_name = ['karan','ravi','ajay','mohan','vipin','kaushal','gaurav']
for i in range(50):
    try:
        classs = random.randint(1,12)
        name = random.choice(full_name)
        phone = random.randint(7898987876,9999999999)
        values = (i+1,classs,name,phone)

        q = f'''insert into python_to_mysql.dbo.mysql_table values{values}'''
        time.sleep(1)
    #     print(q)
        cur.execute(q)
    except Exception as e:
        print(e)
        break
else:
    print('Done')
```
