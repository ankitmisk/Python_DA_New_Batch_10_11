```python
def scrap_meesho(url):
    from selenium import webdriver
    import requests
    from bs4 import BeautifulSoup
    import time
    from datetime import datetime
    import pandas as pd
    import os

#     url = '''https://www.meesho.com/soft-toys-kids/pl/3l1'''
#     url = '''https://www.meesho.com/accessories-men/pl/3tp'''
#     url = '''https://www.meesho.com/home-decor/pl/3tl'''
#     url = '''https://www.meesho.com/bags-ladies/pl/3jo'''

    loc = r"D:\download now\chromedriver-win64\chromedriver-win64\chromedriver.exe"


    cService  = webdriver.ChromeService(executable_path=loc)

    driver = webdriver.Chrome(service = cService)


    driver.get(url)
    data = driver.page_source
    time.sleep(3)

    html_data = data

    driver.close()


    soup = BeautifulSoup(html_data,'html.parser')


    all_class = ['sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA ihwYlH',
                'sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA bGpzcx',
                'sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA jGAjKy',
                'sc-dkrFOg ProductList__GridCol-sc-8lnc8o-0 cokuZA eCJiSA']


    products = []
    prod_price = []
    reviews = []
    rating = []
    prod_offered_price = []
    discount = []

    for cls in all_class:
        for i in soup.find_all('div',{'class':cls}):
            scrap_data = i.text.strip().split('â‚¹')

            if len(scrap_data) == 2:
                prod,rest_details = i.text.strip().split('â‚¹')

                price,rate_rev_details = rest_details.split(' onwardsFree Delivery')


    #             print(prod,price,rate_rev_details)

                products.append(prod)
                prod_price.append(price)

                prod_offered_price.append(price)
                discount.append(0)

                if 'supplier' in rate_rev_details.lower():
                    if len(rate_rev_details) == 0:
                        rating.append(0)
                        reviews.append(0)
                    else:
                        rating.append(rate_rev_details[:3])
                        reviews.append(0)

                elif 'reviews' in rate_rev_details.lower():
                    rat_rv, _ =  rate_rev_details.split(' ')

                    rat = rat_rv[:3]
                    rv = rat_rv[3:]

                    if (len(rat) == 0) or len(rv) == 0:
                        reviews.append(0)
                        rating.append(0)
                    else:
                        reviews.append(rv)
                        rating.append(rat)
                else:
                    rating.append(0)
                    reviews.append(0)

            else:
                prod,offered_price,rest_details = scrap_data

                products.append(prod)
                prod_offered_price.append(offered_price)

                prev_price_dist,rat_rev  =rest_details.split('% offFree Delivery')
                prev_price = prev_price_dist[:3]
                dist = prev_price_dist[3:]
                rat_rv, _ =  rat_rev.split(' ')

                rat = rat_rv[:3]
                rv = rat_rv[3:]

    #             print(rv)
                reviews.append(rv)
                rating.append(rat)

                prod_price.append(prev_price)
                discount.append(dist)





    # print(len(products))
    # print(len(prod_price))
    # print(len(reviews))
    # print(len(rating))
    # print(len(prod_offered_price))
    # print(len(discount))


    all_data = {'Scrap_time':datetime.now(),
                'Product_names':products,
                'Price':prod_price,
                'Offered Price':prod_offered_price,
                'Discount':discount,
                'Rating':rating,
                'Reviews':reviews}

    file_name = url.split('https://www.meesho.com/')[1].split('/')[0]+'.xlsx'
    
    table_new = pd.DataFrame(all_data)
    
    folder_files = os.listdir(r'C:\Users\Lenovo\Desktop\Meesho_scrapped_data')
    
    if file_name in folder_files:
        table_old = pd.read_excel(f'C:\\Users\\Lenovo\\Desktop\\Meesho_scrapped_data\\{file_name}')
        
        
        final_table = pd.concat([table_old,table_new])
        
        final_table.to_excel(f"C:\\Users\\Lenovo\\Desktop\\Meesho_scrapped_data\\{file_name}",index=False)
        
    else:
        table_new.to_excel(f"C:\\Users\\Lenovo\\Desktop\\Meesho_scrapped_data\\{file_name}",index=False)
        
    return table_new
```
