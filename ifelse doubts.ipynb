```python
# Marks

m = Maths = int(input('Maths Marks'))
e = English = int(input('English Marks'))
h = Hindi = int(input('Hindi Marks'))
sst = SocialStudies = int(input('Social Studies Marks'))
s = Science = int(input('Science Marks'))

p = Percentage = ((m+e+h+sst+s)/5)

# is else ladder

if m<40:
    print('You have failed in Maths. You will have to retake this course.')

if e<40:
    print('You have failed in English. You will have to retake this course.')

if h<40:
    print('You have failed in Hindi. You will have to retake this course.')

if sst<40:
    print('You have failed in Social Studies. You will have to retake this course.')

if s<40:
    print('You have failed in Science. You will have to retake this course.')

if 0>m or m>100:
    print('Invalid Marks in Maths')

elif 0>e or e>100:
    print('Invalid Marks in English')

elif 0>h or h>100:
    print('Invalid Marks in Hindi')

elif 0>sst or sst>100:
    print('Invalid Marks in Social Studies')

elif 0>s or s>100:
    print('Invalid Marks in Science')

elif 100>=p>50:
    print(f'You have passes with {p}% marks.')

else:
    print(f'You have scored {p}%. You may have to retake some courses')
```
