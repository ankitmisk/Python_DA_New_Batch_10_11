```python
# magic method  ** imp

class Vector:
    def __init__(self,a,b):
        print('Vector class')
        self.a = a
        self.b = b
        
    def move(self,by):
        self.a += by
        self.b += by
        
        print(f"""vector final points
        a>> {self.a}
        b>> {self.b}""")
        
    def __add__(self,other):
        print(f"""
        a>> {self.a + other.a}
        b>> {self.b + other.b}""")
        
        
    def __sub__(s,o):
        print(f"""
        a>> {s.a - o.a}
        b>> {s.b - o.b}""")
        
    
    def __ge__(self,other):
        print(f"""
        a>> {self.a >= other.a}
        b>> {self.b >= other.b}""")
        
    def __str__(self):
        return 'This is a vector class object!!'
    
#     def __len__(self):
#         return False

    def __len__(self):
        len_a = len(str(self.a))
        len_b = len(str(self.b))
        return  len_a+len_b 
    
    
```
