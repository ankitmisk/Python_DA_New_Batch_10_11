```python
def give_image(name):
    url = '''https://picsum.photos/1080/1080'''
    response = r.get(url)
    data = response.content
    
    with open(f"C:\\Users\\Lenovo\\Desktop\\instagram_pics\\image_{name}.jpeg",'wb') as f:
        f.write(data)
```

```python
give_image('shrey_1')
```

```python
import threading as th
```

```python
t1 = th.Thread(target=give_image,args=['aman'])
```

```python
t1.start()
```

```python
start = time.time()

all_threads = []

for i in range(1,1001):
    t = th.Thread(target=give_image,args=[i])
    all_threads.append(t)
    
for j in all_threads:
    j.start()

end = time.time()

ttt = end - start

print('Total Time Taken: ',ttt)
```

```python
import multiprocessing as mp
```

```python
start = time.time()

all_process = []

for i in range(1,1001):
    t = mp.Process(target=give_image,args=[i])
    all_process.append(t)
    
for j in all_process:
    j.start()

end = time.time()

ttt = end - start

print('Total Time Taken: ',ttt)
```
